```python
import pandas as pd
import requests
```

```python
url = 'https://servicodados.ibge.gov.br/api/v2/censos/nomes/matheus/'

params = {'sexo': 'M',
          'localidade': 32}

response = requests.request("GET", url = url, params = params)
```

```python
# Confere o status
response.status_code
```

```python
# Seleciona o primeiro Ã­ndice da lista
data_json = response.json()[0]['res']
```

```python
# Cria o dataframe
matheus_df = pd.DataFrame(data_json)
display(matheus_df)
```

```python
import seaborn as sns
from matplotlib import pyplot as plt

sns.barplot(data = matheus_df, x = 'periodo', y = 'frequencia')
plt.xticks(rotation = 55);
```
